```python
import pandas as pd
```

```python
## data was obtained by running Serdar's BLS python script - which grabs data nationwide.

monthly_unemployment = pd.read_csv('./data/bls_monthly_unemployment_by_county.csv')
```

```python
monthly_unemployment.head()
```

```python
ca_monthly_unemployment = monthly_unemployment[monthly_unemployment["state"] == "CA"]
```

```python
ca_monthly_unemployment["county"].unique()
```

```python
bay_area_counties = ['Alameda County', 'Contra Costa County', 'Marin County', 'Napa County', 'San Francisco County/city', 
                     'San Mateo County', 'Santa Clara County', 'Solano County', 'Sonoma County']
```

```python
bay_area_unemployment = ca_monthly_unemployment[ca_monthly_unemployment["county"].isin(bay_area_counties)]
```

```python
bay_area_unemployment_map = bay_area_unemployment[bay_area_unemployment["year"] == 2021]
bay_area_unemployment_map = bay_area_unemployment[bay_area_unemployment["month"] == 12]
bay_area_unemployment_map.to_csv('./data/unemployment_bay.csv')
```

```python
bay_area_unemployment = bay_area_unemployment.drop(columns=['fips_state', 'fips_county', 'laus_area_code',
                                                           'civ_labor_force','employed','unemployed', 'area', 
                                                            'state', 'month_name', 'month'])
bay_area_unemployment
```

```python
bay_area_unemployment_2021 = bay_area_unemployment[bay_area_unemployment["year"] == 2021]
bay_area_unemployment_2022 = bay_area_unemployment[bay_area_unemployment["year"] == 2022]
```

```python
bay_area_unemployment_2021.head()
```

```python
pivot_counties_2021 = bay_area_unemployment_2021.pivot(index='date', columns='county', values='unemployed_rate')\
            .reset_index()
pivot_counties_2021.columns.name=None
pivot_counties_2021
```

```python
pivot_counties_2022 = bay_area_unemployment_2022.pivot(index='date', columns='county', values='unemployed_rate')\
            .reset_index()
pivot_counties_2022.columns.name=None
pivot_counties_2022
```

```python
pivot_counties_2021.to_csv('./data/bls_monthly_unemployment_bay_area_2021.csv', index=False)
pivot_counties_2022.to_csv('./data/bls_monthly_unemployment_bay_area_2022.csv', index=False)
```
